```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
from matplotlib.pyplot import figure

plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# reading data from csv file
df = pd.read_csv(r'/Users/saan/Documents/movies.csv.zip')
```

```python
# checking the data

df.head()
```

```python
# checking for any data that is null

for col in df.columns:
    missing_data = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_data))
```

```python
# Data cleaning
```

```python
# Checking the data types

df.dtypes
```

```python
# release year

df['year of release'] = df['year'].astype(str).str[:]
```

```python
# ordering the data by gross revenue

df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# Removing all duplicates in the table

df.drop_duplicates()
```

```python
# Assumptions

# High correlation between budget and gross revenue
# High correlation between company and gross revenue
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

# Naming axes
plt.title('Budget vs Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Film Budget')

# Visualize the scatter plot
plt.show()
```

```python
df.head()
```

```python
# Regression plot for budget vs gross revenue

sns.regplot(x = 'budget', 
            y = 'gross', 
            data = df, 
            scatter_kws = {'color':'red'}, 
            line_kws = {'color':'black'})
```

```python
# This shows positive correlation
```

```python
# Finding correlation

df.corr
```

```python
# Visualizing the correlation matrix

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot = True)

# Naming axes
plt.title('Correlation Matrix')
plt.xlabel('Numeric movie features')
plt.ylabel('Numeric movie features')

plt.show()
```

```python
# Correlation between company and gross revenue

df_numerized = df

for col in df_numerized.columns:
    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
        
        
```
